```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
```

```python
url = "https://es.wikipedia.org/wiki/Anexo:Futbolistas_de_la_selecci%C3%B3n_de_f%C3%BAtbol_de_Espa%C3%B1a"

request = urllib.request.urlopen(url) #Selecting URL to parse
soup = BeautifulSoup(request, "html.parser") #Selecting the parse
```

It is necessary to take a look on the webpage what we want to scrap. It is easier doing `right click > "Inspeccionar"` or `CTRL + MAYUS + I`.

![bs1](img/table_1.png)

As it is under "table" tag, we are going to search all the tables.

```python
all_tables = soup.find_all("table")
all_tables
```

The table we want to get is under the class "sortable", so lets make our search stretcher.>

```python
my_table = soup.find("table", class_="sortable")
```

```python
my_table
```

And following the wikipedia and this, we can see that:
- It has 7 fields
- The data is under "tr" tag and delimited by "td".

```python
#From A to G because we have 7 fields.
A = []
B = []
C = []
D = []
E = []
F = []
G = []

for row in my_table.findAll("tr"):
    cell = row.findAll("td")
    if len(cell)==7: #7 because the table has 7 fields
        A.append(cell[0].find(text=True)) #We are increasing in each case the cell number in the loop
        B.append(cell[1].find(text=True))
        C.append(cell[1].find(text=True))
        D.append(cell[3].find(text=True))
        E.append(cell[4].find(text=True))
        F.append(cell[5].find(text=True))
        G.append(cell[6].find(text=True))
```

```python
F
```

Now we have scrapped the fields. Nonetheless, there's an issue with C fields as it gives back "blank". Let's look it deepler removing "text=True"

```python
A = []
B = []
C = []
D = []
E = []
F = []
G = []

for row in my_table.findAll("tr"):
    cell = row.findAll("td")
    if len(cell)==7:
        A.append(cell[0].find(text=True))
        B.append(cell[1].find(text=True))
        C.append(cell[2])
        D.append(cell[3].find(text=True))
        E.append(cell[4].find(text=True))
        F.append(cell[5].find(text=True))
        G.append(cell[6].find(text=True))
```

```python
C
```

We can see that the names are not under "text" but under "< a href>" tag.
So let's make change this in our code.

```python
A = []
B = []
C = []
D = []
E = []
F = []
G = []

for row in my_table.findAll("tr"):
    cell = row.findAll("td")
    if len(cell)==7:
        A.append(cell[0].find(text=True))
        B.append(cell[1].find(text=True))
        C.append(cell[2].find(href=True))
        D.append(cell[3].find(text=True))
        E.append(cell[4].find(text=True))
        F.append(cell[5].find(text=True))
        G.append(cell[6].find(text=True))
```

```python
C
```

We are closer, but we need to extract the info under href, so let's create other loop:

```python
A = []
B = []
C = []
D = []
E = []
F = []
G = []

for row in my_table.findAll("tr"):
    cell = row.findAll("td")
    if len(cell)==7:
        A.append(cell[0].find(text=True))
        B.append(cell[1].find(text=True))
        for hrefs in cell[2].find(href=True):
            C.append(hrefs)
        D.append(cell[3].find(text=True))
        E.append(cell[4].find(text=True))
        F.append(cell[5].find(text=True))
        G.append(cell[6].find(text=True))
```

Let's create now a dataframe:

```python
spain = {"Jugador" : C,
        "Periodo" : D,
        "Debut" : E,
        "Partidos" : F,
        "Goles" : G}
```

```python
df = pd.DataFrame(spain)
```

```python
df
```

Finally, remove the "\n" in the dataframe

```python
df_clean = df.copy()

for column in df.columns:
    df_clean[column] = df_clean[column].apply(lambda x: x.strip("\n"))
```

```python
df_clean
```

Let's parse the spanish dates

```python
from datetime import datetime as dt
import locale

#Let's set locale as spain to parse the dates
locale.setlocale(locale.LC_ALL, 'esp_esp')

df_clean["parse_debut"] = df_clean["Debut"].apply(lambda x: dt.strptime(x, '%d de %B de %Y'))
df_clean["parse_debut"] = df_clean["parse_debut"].apply(lambda x: x.strftime('%Y-%m-%d'))

df_clean
```

```python
df_clean.to_csv("jug_hist_sp.csv")
```
